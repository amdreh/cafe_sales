```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("/kaggle/input/cafe-sales-dirty-data-for-cleaning-training/dirty_cafe_sales.csv")
```

```python
df=df.drop("Transaction ID",axis=1)
```

```python
for var in df:
    df[var] = df[var].replace(["UNKNOWN", "ERROR", "nan"], np.nan)
    df[var].fillna(df[var].mode()[0],inplace=True)   
```

```python
df.isnull().sum()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
df["year"]=df['Transaction Date'].dt.year
df["year"]
```

```python
df["month"]=df['Transaction Date'].dt.month
df["month"]
```

```python
df["day"]=df['Transaction Date'].dt.day
df["day"]
```

```python
df.drop('Transaction Date',axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
numerical=["Quantity","Price Per Unit","Total Spent"]
for var in df[numerical]:
    #df[var]=df[var].str.replace(".","",regex=True)
    #if you want turn to int 
    df[var]=df[var].astype(float)
```

```python
df.info()
```

```python
cat=[var for var in df.columns if df[var].dtype=="O"]
cat
```

```python
le = LabelEncoder()
for var in df[cat]:
   df[var] = le.fit_transform(df[var])
```
